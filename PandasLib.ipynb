# Pandas Series

```python
import pandas as pd
```

```python
l = [x for x in range(5)]
```

```python
s = pd.Series(l)
s
```

```python
s[3]
```

```python
s = pd.Series(l,index=['a','b','c','d','e'])
s
```

```python
data = {'a':1, 'b':2, 'c':3, 'd':4}
s = pd.Series(data)
s
```

```python
s = pd.Series(data,index=['a','b', 'f'])
s
```

# Creating a series

```python
import pandas as pd
```

```python
s = pd.Series([x for x in range(1,11)])
s
```

```python
s.iloc[0]
```

```python
s.where(s%2==0)
```

```python
s.where(s%2==0, 'Odd')
```

```python
s.where(s%2==0, s**2)
```

```python
s.where(s%2==0, inplace=True)
s.dropna()
```

```python
s.fillna('Filled Value')
```

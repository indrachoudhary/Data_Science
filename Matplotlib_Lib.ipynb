```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from matplotlib import style
%matplotlib inline
```

```python
randomNum = np.random.rand(15)
print (randomNum)
```

```python
#select style of the plot
style.use('ggplot')
plt.plot(randomNum, 'r', label='line one', linewidth=2)
plt.xlabel('Range')
plt.ylabel('Numbers')
plt.title('First plot')

plt.legend()
plt.show
```

```python
cause = 'Chronic diseases','Unintentional injuries', 'Alzheimers','others','flu'
percentile = [62,5,4,8,30]
```

```python
plt.figure(figsize=(5,5))
explode = (0.05,0,0,0,0)
plt.pie(percentile, labels=cause, explode=explode, autopct='%1.1f%%', startangle=70)
plt.axis('equal')
plt.title('Ohio state - 2012 : Leading causes of Deaths ')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
sns.set_theme()
```

```python
fmri = sns.load_dataset("fmri")
```

```python
sns.relplot(
data=fmri,kind="line", x="timepoint", y="signal", col="region", hue="event", style="event")
```

```python
df = pd.read_csv("E:\Data_Science\Data sets\ds_salaries.csv")
```

```python
df.head()
```

```python
df["salary - work_year"] = df.salary - df.work_year
df['remote_ratio'] = df['salary_in_usd'].rolling(100).mean()
sns.set_style('darkgrid')

ax = plt.axes(projection='3d')
ax.scatter(df.index,df["salary - work_year"],df['remote_ratio'])
ax.set_xlabel('Index')
ax.set_ylabel("salary - work_year")
ax.set_zlabel('remote-ratio')
plt.show
```

```python
z1 = np.linspace(0,10,100)
x1 = np.cos(4+z1)
y1 = np.sin(4+z1)
```

```python
sns.set_style('whitegrid')
ax = plt.axes(projection = '3d')
ax.plot3D(x1,y1,z1)
plt.show
```

```python
def return_z(x,y):
    return 50-(x**2+y**2)
```

```python
sns.set_style('whitegrid')
```

```python
x1,y1 = np.linspace(-5,5,50), np.linspace(-5,5,50)
```

```python
x1,y1 = np.meshgrid(x1,y1)
z1 = return_z(x1,y1)
```

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(x1,y1,z1)
plt.show
```
